## Handling Outliers

```python
# Example dataset
data = [5, 2, 3, 4, 7, 1, 6, 1, 1000, 2000, 3000, 4000]

print("Dataset:", data)
```

### IQR

```python
quartiles = np.percentile(data, [25, 50, 75])

# Print results
print("Dataset:", data)
print('\n')
print("25th Percentile (Q1):", quartiles[0])
print("50th Percentile (Median, Q2):", quartiles[1])
print("75th Percentile (Q3):", quartiles[2])
print('\n')
print("Interquartile Range (IQR):", quartiles[2] - quartiles[0])
```

```python
lower_bound = quartiles[0] - 1.5 * (quartiles[2] - quartiles[0])
upper_bound = quartiles[2] + 1.5 * (quartiles[2] - quartiles[0])
```

```python
Lower_Bound
```

```python
Upper_Bound
```

```python
# Find outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]
```

```python
outliers
```

### Z-Score

```python
z_scores = (np.array(data) - np.mean(data)) / np.std(data)
```

```python
data
```

```python
z_scores
```

```python
threshold = 2
```

```python
# Find outliers
outliers = [data[i] for i in range(len(data)) if abs(z_scores[i]) > threshold]

print("Outliers:", outliers)
```

## Data Transformation

### Normalisation and Standardisation

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Sample data for Indian students
data = {
    'Name': ['Rahul', 'Sneha', 'Ravi', 'Anjali', 'Amit'],
    'Age': [20, 21, 19, 22, 20],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
    'City': ['Delhi', 'Mumbai', 'Bangalore', 'Kolkata', 'Chennai'],
    'Marks': [85, 92, 78, 88, 90],
    'Height_cm': [170, 165, 180, 160, 175],
    'Grade': ['A', 'A+', 'B+', 'A', 'A-']
}

# Create DataFrame
df = pd.DataFrame(data)

# Display DataFrame
print("Original DataFrame:")
df
```

```python
# Normalize numeric columns (Age, Marks, Height_cm)

scaler = MinMaxScaler()

df[['Age', 'Marks', 'Height_cm']] = scaler.fit_transform(df[['Age', 'Marks', 'Height_cm']])

# Display DataFrame after normalization
print("\nDataFrame after Normalization:")
df
```

```python
# Standardize numeric columns (Age, Marks, Height_cm)

scaler = StandardScaler()

df[['Age', 'Marks', 'Height_cm']] = scaler.fit_transform(df[['Age', 'Marks', 'Height_cm']])

# Display DataFrame after standardization
print("\nDataFrame after Standardization:")
df
```

### Encoding Categorical Variables

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import category_encoders as ce

# Sample data for Indian students
data = {
    'Name': ['Rahul', 'Sneha', 'Ravi', 'Anjali', 'Amit', 'Siddhant', 'Sheena'],
    'Age': [20, 21, 19, 22, 20, 23, 19],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female'],
    'City': ['Delhi', 'Mumbai', 'Bangalore', 'Kolkata', 'Chennai', 'Delhi', 'Mumbai'],
    'Marks': [85, 92, 78, 88, 90, 54, 63],
    'Height_cm': [170, 165, 180, 160, 175, 172, 169],
    'Grade': ['A', 'A+', 'B+', 'A', 'A-', 'B-', 'A-'],
    'Staus':[1, 1, 1, 0, 0, 1,1]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display DataFrame
print("Original DataFrame:")
df
```

```python
# Label Encoding for Gender column
le = LabelEncoder()
df['Gender_LabelEncoded'] = le.fit_transform(df['Gender'])

# Display DataFrame after Label Encoding for Gender
print("\nDataFrame after Label Encoding for Gender:")
df
```

```python
# One-Hot Encoding for City column
ohe = OneHotEncoder(sparse=False)
ohe_result = ohe.fit_transform(df[['City']])
df_ohe = pd.DataFrame(ohe_result, columns=ohe.get_feature_names(['City']))

# Concatenate the original DataFrame with the one-hot encoded DataFrame
df = pd.concat([df, df_ohe], axis=1)

# Drop the original 'City' column
df.drop(columns=['City'], inplace=True)

# Display DataFrame after One-Hot Encoding for City
print("\nDataFrame after One-Hot Encoding for City:")
df
```

```python
# Ordinal Encoding for Grade column
ordinal_encoder = ce.OrdinalEncoder(cols=['Grade'], mapping=[{'col': 'Grade', 'mapping': {'B-': 5, 'A-': 3, 'B+': 4, 'A': 2, 'A+': 1}}])
df = ordinal_encoder.fit_transform(df)

# Display DataFrame after Ordinal Encoding for Grade
print("\nDataFrame after Ordinal Encoding for Grade:")
df
```

```python
# Target Encoding for Name column
target_encoder = ce.TargetEncoder(cols=['Gender'])
df = target_encoder.fit_transform(df, df['Staus'])

# Display DataFrame after Target Encoding for Name
print("\nDataFrame after Target Encoding for Gender:")
df
```

```python
data = {
    'Name': ['Rahul', 'Sneha', 'Ravi', 'Anjali', 'Amit', 'Siddhant', 'Sheena'],
    'Age': [20, 21, 19, 22, 20, 23, 19],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female'],
    'City': ['Delhi', 'Mumbai', 'Bangalore', 'Kolkata', 'Chennai', 'Delhi', 'Mumbai'],
    'Marks': [85, 92, 78, 88, 90, 54, 63],
    'Height_cm': [170, 165, 180, 160, 175, 172, 169],
    'Grade': ['A', 'A+', 'B+', 'A', 'A-', 'B-', 'A-'],
    'Staus':[1, 1, 1, 0, 0, 1,1]
}

# Create DataFrame
df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt

# Plot histograms for Age, Marks, and Height_cm columns
plt.figure(figsize=(10, 6))

# Histogram for Age
plt.subplot(2, 2, 1)
plt.hist(df['Age'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Histogram for Marks
plt.subplot(2, 2, 2)
plt.hist(df['Marks'], bins=4, color='salmon', edgecolor='black')
plt.title('Histogram of Marks')
plt.xlabel('Marks')
plt.ylabel('Frequency')

# Histogram for Height_cm
plt.subplot(2, 2, 3)
plt.hist(df['Height_cm'], bins=3, color='lightgreen', edgecolor='black')
plt.title('Histogram of Height_cm')
plt.xlabel('Height_cm')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
